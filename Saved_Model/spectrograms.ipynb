```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram

# sampling frequency
fs = 173.61

input_dir = "Base_Dataset"
output_dir = "spectrogram_dataset"

classes = {
    "Z": "healthy",
    "O": "healthy",
    "N": "interictal",
    "F": "interictal",
    "S": "ictal"
}

for folder in classes:
    
    class_name = classes[folder]
    input_path = os.path.join(input_dir, folder)
    output_path = os.path.join(output_dir, class_name)

    os.makedirs(output_path, exist_ok=True)

    for file in os.listdir(input_path):
        if not file.lower().endswith(".txt"):
            continue

        file_path = os.path.join(input_path, file)

        # load EEG signal
        signal = np.loadtxt(file_path)

        # compute spectrogram
        f, t, Sxx = spectrogram(signal, fs)

        plt.figure(figsize=(3,3))
        #plt.pcolormesh(t, f, Sxx, shading='gouraud')
        plt.pcolormesh(t, f, 10*np.log10(Sxx), shading='gouraud')
        plt.axis('off')

        #save_name = file.replace(".txt", ".png")
        save_name = os.path.splitext(file)[0] + ".png"
        save_path = os.path.join(output_path, save_name)

        plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
        
        plt.close()

print("Spectrogram dataset created.")
```
