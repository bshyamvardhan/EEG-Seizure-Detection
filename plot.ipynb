**Plotting the signals for all the 3 classes**

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load CSVs
ictal_df = pd.read_csv("csv_output/ictal.csv")
nonictal_df = pd.read_csv("csv_output/non_ictal.csv")
healthy_df = pd.read_csv("csv_output/healthy.csv")

def plot_signal(df, label, samples=4096):
    signal = df.iloc[:samples]
    plt.figure(figsize=(10, 3))
    plt.plot(signal["time_sec"], signal["amplitude"])
    plt.title(f"EEG Signal - {label}")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

# Plot one signal from each class
plot_signal(ictal_df, "Ictal")
plot_signal(nonictal_df, "Non-Ictal")
plot_signal(healthy_df, "Healthy")
```

```python
def plot_multiple(df, label, n_signals=5):
    SAMPLES=4096
    plt.figure(figsize=(10, 4))
    for i in range(n_signals):
        sig = df.iloc[i*SAMPLES:(i+1)*SAMPLES]
        plt.plot(sig["time_sec"], sig["amplitude"], alpha=0.6)
    plt.title(f"{label} - {n_signals} EEG signals")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Amplitude")
    plt.show()

plot_multiple(ictal_df, "Ictal")
plot_multiple(nonictal_df, "Non-Ictal")
plot_multiple(healthy_df, "Healthy")
```

This confirms:

No corrupted signals

All files same length

No mislabeling
